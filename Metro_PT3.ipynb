```python
data = pd.DataFrame(data, columns=col_name)
```

```python
# Creating a New Column Shape to Check if is there any anomaly
from sklearn.ensemble import IsolationForest
model = IsolationForest(contamination='auto')
model.fit(data)
anomaly_scores = model.predict(data)
```

```python
isFailure = pd.DataFrame(anomaly_scores, columns=["Failure"])
isFailure = isFailure.replace(-1, 0)
isFailure.head()
```

```python
df = pd.concat([data, isFailure], axis=1)
df.head()
```

```python
failure = (isFailure["Failure"] == 0).value_counts()
print(failure)
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, random_state = 42)
```

```python
train_labels = train["Failure"]
test_labels = test["Failure"]
```

```python
print(train.shape)
print(test.shape)
print(train_labels.shape)
print(test_labels.shape)
```

```python
train = train.drop(columns=["Failure"], axis=1)
test = test.drop(columns=["Failure"], axis=1)
print(train.shape)
print(test.shape)
```

```python
input_data = np.asarray(train)
```

```python
input_data
```

```python
input_data.shape
```

```python
input_test_data = np.asarray(test)
```

```python
import time
```

```python
from sklearn.linear_model import LogisticRegression
LRModel = LogisticRegression()
s = time.time()
LRModel.fit(train, train_labels)
e = time.time()
print("Execution Time", e-s)
LRPred = LRModel.predict(input_data)
LR_test_pred = LRModel.predict(test)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFCModel = RandomForestClassifier()
s = time.time()
RFCModel.fit(train, train_labels)
e = time.time()
print("Execution Time", e-s)
RFC_train_pred = RFCModel.predict(input_data)
RFC_test_pred = RFCModel.predict(input_test_data)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

cm_LR = confusion_matrix(test_labels, LR_test_pred)
cm_RFC = confusion_matrix(test_labels, RFC_test_pred)
cm_lgbm = confusion_matrix(test_labels, lgbm_test_pred)
cm_catb = confusion_matrix(test_labels, catb_test_pred)
cm_xgb = confusion_matrix(test_labels, xgb_test_pred)

accuracy_LR = accuracy_score(test_labels, LR_test_pred)
accuracy_RFC = accuracy_score(test_labels, RFC_test_pred)
accuracy_lgbm = accuracy_score(test_labels, lgbm_test_pred)
accuracy_catb = accuracy_score(test_labels, catb_test_pred)
accuracy_xgb = accuracy_score(test_labels, xgb_test_pred)

precision_LR = precision_score(test_labels, LR_test_pred)
precision_RFC = precision_score(test_labels, RFC_test_pred)
precision_lgbm = precision_score(test_labels, lgbm_test_pred)
precision_catb = precision_score(test_labels, catb_test_pred)
precision_xgb = precision_score(test_labels, xgb_test_pred)

recall_LR = recall_score(test_labels, LR_test_pred)
recall_RFC = recall_score(test_labels, RFC_test_pred)
recall_lgbm = recall_score(test_labels, lgbm_test_pred)
recall_catb = recall_score(test_labels, catb_test_pred)
recall_xgb = recall_score(test_labels, xgb_test_pred)

print("Logistic Regression:\n")
print("Confusion Matrix:\n\n", cm_LR)
print("Accuracy:", accuracy_LR)
print("Precision:", precision_LR)
print("Recall:", recall_LR)

print("\nRandom Forest:")
print("Confusion Matrix:\n", cm_RFC)
print("Accuracy:", accuracy_RFC)
print("Precision:", precision_RFC)
print("Recall:", recall_RFC)

print("\nLightGBM:")
print("Confusion Matrix:\n", cm_lgbm)
print("Accuracy:", accuracy_lgbm)
print("Precision:", precision_lgbm)
print("Recall:", recall_lgbm)

print("\nCatBoost:")
print("Confusion Matrix:\n", cm_catb)
print("Accuracy:", accuracy_catb)
print("Precision:", precision_catb)
print("Recall:", recall_catb)

print("\nXGBoost:")
print("Confusion Matrix:\n", cm_xgb)
print("Accuracy:", accuracy_xgb)
print("Precision:", precision_xgb)
print("Recall:", recall_xgb)
```
